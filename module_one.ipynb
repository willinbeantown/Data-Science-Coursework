```python
import math
import pandas
import random
```

```python
data = {
    "names": ["George", "John", "Thomas", "James", "James", "John"],
    "ages": [67, 90, 83, 85, 73, 80]
}

population_df = pandas.DataFrame(data)
```

```python
population_df
```

```python
sample_ids = []
used_i = []

while len(sample_ids) < 3:
    rando = random.randint(1, len(population_df))
    if not rando in used_i:
        used_i.append(rando)
        sample_ids.append(rando-1)
    
sample_df = population_df.iloc[sample_ids]
```

```python
sample_df
```

```python
mean = sum(population_df["ages"].to_list())/len(population_df)

print(mean)
```

```python
standard_dev = math.sqrt(1/len(population_df)*mean)

print(standard_dev)
```

```python
variance = standard_dev ** 2  # the exponentiation operator

print(variance)
```

```python
# If sigma is unknown, but the 95 percentile (aka 2sigma intercept) value IS known
two_sigma_intercept = mean+(2*standard_dev)  # a point along the curve

sigma = (two_sigma_intercept-mean)/2

print(sigma)
```
