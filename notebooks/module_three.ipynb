```python
cursor.execute("""
CREATE TABLE qctestdb.testresults (
  plant INT NOT NULL PRIMARY KEY,
  testsults VARCHAR(255) NOT NULL,
  zip INT NOT NULL)
""")

db_connection.commit()  # REQUIRED TO MAKE UPDATES
```

```python
# Confirm Table and Fields Created.
cursor.execute("DESCRIBE qctestdb.testresults")

for row in cursor:

    print(row)
```

```python
data = [
    (1001, "B1234PABQ50D20230908", 93722),
    (1002, "B1234PABQ0050D20230908B5678PXYQ0030D20230909", 1960),
    (1006, "B1234PABQ0050D20230908", 2703),
    (1007, "B2021PGEQ25D20230103", 77433),
    (1008, "B4521PABQ56D20230315", 33023)
]
```

```python
insert_sql = """
INSERT INTO qctestdb.testresults (plant, testsults, zip)
VALUES (%s, %s, %s)
"""  # `%s` is a placeholder for some string.
# use of `%s` is overwritten by database type.

cursor.executemany(insert_sql, data)

db_connection.commit()

print(f"{cursor.rowcount} rows inserted")
```

# Search the DB

```python
cursor.execute("SELECT * FROM qctestdb.testresults")

rows = cursor.fetchall()

for row in rows:
    
    print(row)
```

```python
cursor.execute("SELECT testsults, zip FROM qctestdb.testresults WHERE plant = 1001")

rows = cursor.fetchall()

for row in rows:
    
    print(row)
```
