### INTRODUCTION TO STATISTICAL TERMINOLOGY

A small iPython notebook using `math`, `pandas`, and `random` to introduce statistical formula for **Mean**, **Standard Deviation**, **Sample Standard Deviation**, **Variance**, _ _ _ , and to practice Python in a Jupyter Notebook.

**PART 1**

Create a data table of a population.

For this example, we use all first names and ages of Presidents
of the United States born in the 1700s.

```python
from pandas import DataFrame

data = {
    "names": ["George", "John", "Thomas", "James", "James", "John",
              "Andrew", "Martin", "William", "John", "James", "Zachary",
              "James"],
    "ages": [67, 90, 83, 85, 73, 80,
             78, 79, 68, 71, 53, 65,
             77]
}

population_df = DataFrame(data)

population_df
```

**PART 2**

Create a sample from the above data by selecting 5 random Presidents.

```python
from random import randint

sample_ids = []
used_i = []

while len(sample_ids) <= 5:
    rando = randint(1, len(population_df))
    if not rando in used_i:
        used_i.append(rando)
        sample_ids.append(rando-1)
    
sample_df = population_df.iloc[sample_ids]

sample_df
```

**PART 3**

With the Population defined above and built-in Python functions, create formula for **Mean**, **Standard Deviation**, **Variance** and **sigma**

```python
mean = sum(population_df["ages"].to_list())/len(population_df)

mean
```

```python
from math import sqrt

standard_dev = sqrt(1/len(population_df)*mean)

standard_dev
```

```python
variance = standard_dev ** 2

variance
```

```python
# If sigma is unknown, but the 95 percentile (aka 2sigma intercept) value IS known
two_sigma_intercept = mean+(2*standard_dev)  # a point along the curve

sigma = (two_sigma_intercept-mean)/2

sigma
```
