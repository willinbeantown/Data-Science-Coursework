```python
temperature = [40, 50, 60, 70, 90, 105]

# Old Way
# tempaverage = (temperature[0] + temperature[1] + ...)/number_of_temperatures)
tempaverage = sum(temperature)/len(temperature)

print(tempaverage)
```

```python
import statistics
```

```python
tempmean = statistics.mean(temperature)
print(f"Mean = {tempmean}")
```

```python
tempmedian = statistics.median(temperature)
print(tempmedian)
```

```python
grades = [70, 70, 80, 80, 90, 100]
grademode = statistics.mode(grades)
print(grademode) # HAS MULTIPLE MODES THOUGH!!
```

```python
from matplotlib import pyplot

temp = [40, 50, 60, 70, 80, 90, 100]
customer = [45, 70, 90, 180, 250, 300, 60]

pyplot.plot(temp, customer)
pyplot.title("Temp vs. Customers")
pyplot.xlabel("Temperature (in F)")
pyplot.ylabel("Number of Customers")
pyplot.show()
```
